```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
car = cv2.imread("C:\\Users\\USER\\Desktop\\opencv\\DATA\\car_plate.jpg")
```

```python
def display(img):
    # fill me in!
    fig = plt.figure(figsize=(10,8))
    ax =fig.add_subplot(111)
    new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
n_plate_cascade = cv2.CascadeClassifier("C:\\Users\\USER\\Desktop\\opencv\\DATA\\haarcascades\\haarcascade_russian_plate_number.xml")
```

```python
def detect_plate(img):
    new_img_casc = img.copy()
    casc = n_plate_cascade.detectMultiScale(new_img_casc,scaleFactor=1.2,minNeighbors=3)
    for (x,y,w,h) in casc:
        cv2.rectangle(new_img_casc,(x,y),(x+w,y+h),(0,0,255),3)
    return new_img_casc
```

```python
result = detect_plate(car)
```

```python
display(result)
```

```python
def detect_and_blur_plate(img):
    
    # fill me in
    plate_img = img.copy()
    roi = img.copy()
    casc = n_plate_cascade.detectMultiScale(plate_img,scaleFactor=1.2,minNeighbors=3)
    for (x,y,w,h) in casc:
        roi = roi[y:y+h,x:x+w]
        my_blurred = cv2.medianBlur(roi,7)
        plate_img[y:y+h,x:x+w]=my_blurred
    return plate_img
```

```python
result = detect_and_blur_plate(car)
```

```python
display(result)
```
